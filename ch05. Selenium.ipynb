#Selenium으로 실습하기

###동적 웹 페이지 불러오기

```python
from selenium import webdriver

# 동적 웹페이지 크롤링을 위한 웹 드라이버 사용
driver = webdriver.Chrome("C:/Users/liver/Documents/chromedriver_win32/chromedriver")

driver.implicitly_wait(3)   # 페이지 로드에 3초 기다림
driver.get("https://www.livesport.com/kr/team/liverpool/lId4TMwf/")   # 해당 페이지 불러오기
```

```python
from bs4 import BeautifulSoup

page = driver.page_source    # html코드 저장해주는 page_source 함수
bs_obj = BeautifulSoup(page, "html.parser")

# 이제 div 태그도 찾아낸다
print(bs_obj)
```

**승패 기록 페이지 크롤링하고 분석하기**

최근 점수에는 이전 10경기의 결과가 나옴

```python
# 승리 기록
win = bs_obj.find_all("span", {"class" : "wld wld--w"})
print(win)
print(len(win))
```

```python
# 무승부였다가 정규시간 이후에 승리한 것도 체크

tiewin = bs_obj.find_all("span", {"class" : "wld wld--wo"})
print(tiewin)
print(len(tiewin))
```

```python
# 비긴 경기 없음
draw = bs_obj.find_all("span", {"class" : "wld wld--d"})
print(draw)
print(len(draw))
```

```python
# 진 경기 없음
lose = bs_obj.find_all("span", {"class" : "wld wld--l"})
print(lose)
print(len(lose))
```

```python
# 무승부였다가 정규시간 이후에 패배
tielose = bs_obj.find_all("span", {"class" : "wld wld--lo"})
print(tielose)
print(len(tielose))
```

```python
n_win = len(win) + len(tiewin)
n_draw = len(draw)
n_lose = len(lose) + len(tielose)

print(n_win, n_draw, n_lose)
```

```python
result = {"win" : n_win, "draw" : n_draw, "lose" : n_lose}

print(result)
```
